# Step 1: Load Data

```python
import pandas as pd
import numpy as np

data = {
    "student": ["A","B","C","A","B","C"],
    "topic": ["Algebra","Algebra","Geometry","Geometry","Statistics","Statistics"],
    "score": [70,85,60,75,90,65],
    "time_spent": [30,25,40,35,20,45]
}

df = pd.DataFrame(data)
```

# Step 2: Analyze Performance

```python
# average score per topic
df.groupby("topic")["score"].mean()

# weakest topic
df.groupby("topic")["score"].mean().sort_values()
```

# Step 3: Visualize Results

```python
import matplotlib.pyplot as plt

df.groupby("topic")["score"].mean().plot(kind="bar")
plt.title("Average Score per Topic")
plt.show()
```

# Step 4: AI Tutor

```python
def ai_tutor(question):
    return f"Explanation for: {question}\nThis is a simplified explanation based on core concepts."
```

# Step 5: Recommendations

```python
def recommend(df):
    weak_topic = df.groupby("topic")["score"].mean().idxmin()
    return f"Focus more on {weak_topic} to improve performance."
```
